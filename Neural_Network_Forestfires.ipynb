```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
data["size_category"]=label.fit_transform(data["size_category"])
data["month"] = label.fit_transform(data["month"])
data["day"] = label.fit_transform(data["day"])
```

```python
X = data[['month',	'day',	'FFMC',	'DMC', 'DC',	'ISI', 'temp', 'RH',	'wind',	'rain']]
y = data['size_category']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(hidden_layer_sizes=(6,5), random_state=3, verbose=True)
clf
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
